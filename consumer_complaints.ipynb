# EDA and Feature Engineering

```python
train_data = df_json.copy()
```

```python
train_data.head()
```

```python
train_data.head(2).T
```

```python
new_train_data = train_data[['product', 'complaint_what_happened']].copy()
new_train_data = new_train_data[pd.notnull(new_train_data['complaint_what_happened'])]
new_train_data.columns = ['product', 'consumer_complaint']
new_train_data
```

```python
new_train_data.shape
```

### percentage complaint with text

```python
total = new_train_data['consumer_complaint'].notnull().sum()
print(total)
round((total/len(train_data)*100),1)
```

### There is no null value on complaint, This is still a good number to work with. Now let's have a look at the categories we want to classify each complaint.

```python
pd.DataFrame(train_data['product'].unique()).values
```

### there are 18 types of categories in product columns

```python
#let's take a sample data, because millions of data are time consuming
sample_data = new_train_data.sample(10000, random_state=1).copy()
```

```python
sample_data.head()
```

```python
new_train_data['product'].unique()
```

```python
sample_data.replace({
    'product': {
        'Credit reporting, credit repair services, or other personal consumer reports': "Credit reporting, repair and other",
        'Credit reporting': 'Credit reporting, repair and other',
        'Credit card': 'Credit card or Prepaid card',
        'Prepaid card': 'Credit card or Prepaid card',
        'Payday loan': 'Payday loan, title loan or personal loan',
        'Money transfers': 'Money transfer, virtual currency or money service',
        'Virtual currency': 'Money transfer, virtual currency or money servic'
    }
}, inplace=True)
```

```python
pd.DataFrame(sample_data['product'].unique())
```

### lets create each category as a number

```python
sample_data['category_id'] = sample_data['product'].factorize()[0]
category_id_sample_data = sample_data[['product', 'category_id']].drop_duplicates()

category_to_id = dict(category_id_sample_data.values)
id_to_category = dict(category_id_sample_data[['category_id','product']].values)

sample_data.head(20)
```

### Let's show a bar chart using the number of complaints

```python
fig = plt.figure(figsize=(8, 6))
colors = ['gray', 'blue', 'green', 'red', 'black', 'azure', 'yellow', 'orange', 'brown', 'tomato', 'darkred', 'gold', 'teal', 'salmon']
sample_data.groupby('product').consumer_complaint.count().sort_values().plot.barh(ylim=0, color=colors,
                                                                                 title='Number of complaints of each catagory\n')
plt.xlabel('Number of occurences', fontsize=10)
```
